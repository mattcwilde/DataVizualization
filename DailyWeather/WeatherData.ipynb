# Exploring the NOAA  National Weather Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
df = pd.read_csv('weather.csv')
```

```python
df[:5]
```

```python
df.plot(x='longitude', y='latitude', kind='scatter', xlim=(-170,-60))
```

```python
df.plot.hexbin(x='longitude', y='latitude',xlim=(-160, -40), figsize=(12*1.33,12))
```

```python
wash_data = df[df['state']=='WA'].copy()
wash_data.set_index("station", inplace=True)
wash_data[:5]
```

```python
wash_data.plot.hexbin(x='longitude', y='latitude')
```

```python
wash_data.shape
```

```python
mean_station = wash_data.groupby(wash_data.index).mean()
```

```python
mean_station.sort_values('TMAX')
```
